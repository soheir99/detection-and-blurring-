```python
import tkinter
from tkinter import *
from tkinter import font 
from tkinter import ttk 
import tkinter as tk 
from PIL import ImageTk,Image
from tkinter import filedialog
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt 


def select_vedio():
 
    fln=filedialog.askopenfilename(initialdir=os.getcwd(),title="Select vedio File ",
                                   filetypes=(("mp4 File","*.mp4"),("All file","*.*")))
    global vedio
    vedio =fln
    print(vedio)
    form.listbox.insert(0,vedio)

def showimage():
 
    fln=filedialog.askopenfilename(initialdir=os.getcwd(),title="Select Image File ",
                                   filetypes=(("JPG File","*.jpg"),("PNG File","*.png"),("All file","*.*")))
    img=Image.open(fln)
    img.thumbnail((250,230))
    img=ImageTk.PhotoImage(img)
    global txtimg2
    txtimg2 =fln
    form.labelHead7.configure(image=img)
    form.labelHead7.image=img
        
def showimage1():
 
    fln1=filedialog.askopenfilename(initialdir=os.getcwd(),title="Select Image File ",
                                   filetypes=(("JPG File","*.jpg"),("PNG File","*.png"),("All file","*.*")))
    img=Image.open(fln1)
    img.thumbnail((220,230))
    img=ImageTk.PhotoImage(img)
    global txtimg
    txtimg =fln1
    print(txtimg)
    form.labelHead4.configure(image=img)
    form.labelHead4.image=img
    
def detect():
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    img = cv2.imread(txtimg2)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

    cv2.imwrite('C:/Users/20155/fff.jpg',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    im=Image.open('C:/Users/20155/fff.jpg')
    im.thumbnail((250,230))
    im=ImageTk.PhotoImage(im)
    form.labelHead8.configure(image=im)
    form.labelHead8.image=im
    
    
def blur():
   
    image = cv2.imread(txtimg) 

    face_detect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 
    face_data = face_detect.detectMultiScale(image, 1.3, 5) 

    for (x, y, w, h) in face_data: 
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2) 
        roi = image[y:y+h, x:x+w] 
        roi = cv2.GaussianBlur(roi, (23, 23), 30)  
        image[y:y+roi.shape[0], x:x+roi.shape[1]] = roi 

    cv2.imwrite('C:/Users/20155/kkk.jpg',image)
    im=Image.open('C:/Users/20155/kkk.jpg')
    im.thumbnail((250,230))
    im=ImageTk.PhotoImage(im)
    form.labelHead5.configure(image=im)
    form.labelHead5.image=im
        
def detect_vedio():
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

    cap = cv2.VideoCapture(vedio)

    while 1:
        ret, img = cap.read()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)

        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]

            eyes = eye_cascade.detectMultiScale(roi_gray)
            for (ex,ey,ew,eh) in eyes:
                cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

        cv2.imshow('img',img)
        k = cv2.waitKey(30) & 0xff
        if k == 27:
            break

    cap.release()
    cv2.destroyAllWindows()

def blur_vedio():
    faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(vedio)
    while True:
        check,img = cap.read()
        faces = faceCascade.detectMultiScale(img,1.1,3)
        for (x, y, w, h) in faces:
            ROI = img[y:y+h, x:x+w]
            blur = cv2.GaussianBlur(ROI, (91,91),0) 
            img[y:y+h, x:x+w] = blur
        if faces==():
            cv2.putText(img,'No Face Found!',(20,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255))
        cv2.imshow('Face Blur',img)
        if cv2.waitKey(1) & 0xff==ord('q'):
            break    
    cap.release()
    cv2.destroyAllWindows()

    
    
fnt='None 25 bold'
fg = '#000000'
bg = '#ffffff'
bgtxt='#ffffff'
pad=10

form = Tk() 
form.title("detecting and bluring faces") 
form.configure(width = 800,height = 600,background = 'lightblue')
w =800
h=600
sw=form.winfo_screenwidth()
sh=form.winfo_screenheight()

xx=(sw-w)/2
yy=(sh-h)/2
form.geometry('%dx%d+%d+%d'%(w,h,xx,yy))
form.resizable(width = False,height = False) 

form.labelHead1 = Label(form,bg = "blue",pady = 5) 
form.labelHead1.place(relwidth = 1,relheight=0.40,rely = 0.40) 


form.labelHead4 = Label(form.labelHead1,bg = "dark cyan",pady = 5) 
form.labelHead4.place(relwidth = 0.40,relheight=1)
 

form.labelHead5 = Label(form.labelHead1,bg = "dark cyan",pady = 5) 
form.labelHead5.place(relwidth = 0.40,relheight=1,relx = 0.40)

form.labelHead6 = Label(form.labelHead1,bg = "gray",pady = 5) 
form.labelHead6.place(relwidth = 0.20,relheight=1,relx = 0.80)

form.button1 = Button(form.labelHead6,bg='lightblue',fg='blue',font = "Helvetica 13 bold",text = 'show',command= showimage1)
form.button1.place(rely=0.15,relx=0.10,width=115,relheight=0.26)
form.button2 = Button(form.labelHead6,bg='lightblue',fg='blue',font = "Helvetica 13 bold",text = 'blur',command= blur)
form.button2.place(rely=0.60,relx=0.10,width=115,relheight=0.26)


form.labelHead2 = Label(form,bg = "cyan",pady = 5) 
form.labelHead2.place(relwidth = 1,relheight=0.10,rely = 0.80) 

form.listbox = Listbox(form.labelHead2,height =2, width =30)
form.listbox.place(rely=0.15,relx=0.05)
form.button6 = Button(form.labelHead2,bg='lightblue',fg='red',font = "Helvetica 13 bold",text = 'select vedio',command=select_vedio)
form.button6.place(rely=0.15,relx=0.30,width=130,height=40)
form.button7 = Button(form.labelHead2,bg='lightblue',fg='red',font = "Helvetica 13 bold",text = 'detect vedio',command=detect_vedio)
form.button7.place(rely=0.15,relx=0.60,width=130,height=40)
form.button8 = Button(form.labelHead2,bg='lightblue',fg='red',font = "Helvetica 13 bold",text = 'blur vedio',command=blur_vedio)
form.button8.place(rely=0.15,relx=0.80,width=130,height=40)


form.labelHead10 = Label(form,bg = "gray",pady = 5) 
form.labelHead10.place(relwidth = 1,relheight=0.10,rely = 0.90) 

form.button5 = Button(form.labelHead10,bg='lightblue',fg='red',font = "Helvetica 13 bold",text = 'EXIT',command=form.destroy)
form.button5.place(rely=0.15,relx=0.45,width=120,height=40)




form.labelHead3 = Label(form,bg = "blue",pady = 5) 
form.labelHead3.place(relwidth = 1,relheight=0.40) 


form.labelHead7 = Label(form.labelHead3,bg = "dark cyan",pady = 5) 
form.labelHead7.place(relwidth = 0.40,relheight=1)

form.labelHead8 = Label(form.labelHead3,bg = "dark cyan",pady = 5) 
form.labelHead8.place(relwidth = 0.40,relheight=1,relx = 0.40)

form.labelHead9 = Label(form.labelHead3,bg = "gray",pady = 5) 
form.labelHead9.place(relwidth = 0.20,relheight=1,relx = 0.80)

form.button3 = Button(form.labelHead9,bg='lightblue',fg='blue',font = "Helvetica 13 bold",text = 'show',command= showimage)
form.button3.place(rely=0.15,relx=0.10,width=115,relheight=0.26)
form.button3 = Button(form.labelHead9,bg='lightblue',fg='blue',font = "Helvetica 13 bold",text = 'detect',command = detect)
form.button3.place(rely=0.60,relx=0.10,width=115,relheight=0.26)



form.mainloop()

```
